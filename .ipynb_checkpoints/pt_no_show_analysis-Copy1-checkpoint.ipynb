```python
# A few of the column names need to be corrected eithe due to spelling or convenience

df = df.rename(columns={'Hipertension':'Hypertension',
                        'No-show':'Noshow',
                        'Handcap':'Handicap'})
df.head()
```

```python
# The time series data need to be standardized for ease of use and we will create a day of the week column for later use

df['ScheduledDay'] = df['ScheduledDay'].apply(np.datetime64)
df['AppointmentDay'] = df['AppointmentDay'].apply(np.datetime64)

# Parse the day of the week from the appointment date

df['DayofTheWeek'] = df['AppointmentDay'].dt.weekday_name

print(df['ScheduledDay'].head())
print(df['AppointmentDay'].head())
print(df['DayofTheWeek'].head())
```

## Adding Variables

```python
# Waiting period calculation

df['WaitPeriod'] = (df['AppointmentDay'] - df['ScheduledDay']) / (np.timedelta64(1, 'D'))
```

```python
# In order to explore the time of day appointments are made the hour of the day will be parsed out from the scheudling column

def calculateHour(timestamp):
    timestamp = str(timestamp)
    hour = int(timestamp[11:13])
    minute = int(timestamp[14:16])
    second = int(timestamp[17:])
    return round(hour + minute/60 + second/3600)

df['HourOfTheDay'] = df.ScheduledDay.apply(calculateHour)
print(df['HourOfTheDay'].head())
```

```python
# Associating probabilities

def probAssos(data, group_by):
    df = pd.crosstab(index = data[group_by], columns = data['Noshow']).reset_index()
    df['probShowUp'] = df['Yes'] / (df['Yes'] + df['No'])
    return df[[group_by, 'probShowUp']]

sns.lmplot(data = probAssos(df, 'Age'), x = 'Age', y = 'probShowUp', fit_reg = True)
plt.xlim(0, 100)
plt.title('Probability of showing up with respect to Age')
plt.show()
```

```python
sns.lmplot(data = probAssos(df, 'HourOfTheDay'), x = 'HourOfTheDay', 
           y = 'probShowUp', fit_reg = True)
plt.title('Probability of showing up with respect to HourOfTheDay')
plt.show()

```

```python
sns.lmplot(data = probAssos(df, 'WaitPeriod'), x = 'WaitPeriod', 
           y = 'probShowUp', fit_reg = True)
plt.title('Probability of showing up with respect to AwaitingTime')
plt.ylim(0, 1)
plt.show()
```

```python
def probAssosbyCat(group_by):
    rows = []
    for item in group_by:
        for level in df[item].unique():
            row = {'Condition': item}
            total = len(df[df[item] == level])
            n = len(df[(df[item] == level) & (df['Noshow'] == 'No')])
            row.update({'Level': level, 'Probability': n / total})
            rows.append(row)
    return pd.DataFrame(rows)
```

```python
sns.barplot(data = probAssosbyCat(['Diabetes', 'Alcoholism', 'Hypertension',
                                          'Scholarship']),
            x = 'Condition', y = 'Probability', hue = 'Level', palette="mako")
plt.title('Probability of showing up')
plt.ylabel('Probability')
plt.show()
```

```python
def posteriorNoShow(condition):
    levels = list(df[condition].unique())
    if condition not in ['DayofTheWeek', 'Gender']: 
        levels.remove(0)
    rows = []
    for level in levels:
        p = len(df[df[condition] == level]) / len(df)
        p1 = len(df[(df[condition] == level) & (df['Noshow'] == 'Yes')]) / len(df[df['Noshow'] == 'Yes'])
        p2 = len(df[(df[condition] == level) & (df['Noshow'] == 'No')]) / len(df[df['Noshow'] == 'No'])
        if len(levels) > 1:
            rows.append({'Levels': level, 
                         'Probability': (p * p1) / (p * p1 + p * p2)})
        else:
            rows.append({'Condition': condition,
                         'Probability': (p * p1) / (p * p1 + p * p2)})
    return rows
    
tuples = []
tuples.extend(posteriorNoShow('Diabetes'))
tuples.extend(posteriorNoShow('Hypertension'))
tuples.extend(posteriorNoShow('Alcoholism'))
tuples.extend(posteriorNoShow('SMS_received'))
tuples.extend(posteriorNoShow('Scholarship'))

sns.barplot(data = pd.DataFrame(tuples)[['Condition', 'Probability']], 
            x = 'Condition', y = 'Probability', palette="mako")
plt.title('Posterior probability of diseases and scholarship given a no-show')
plt.ylabel('Probability')
plt.show()
```

```python
sns.barplot(data = pd.DataFrame(posteriorNoShow('Handicap')), 
            x = 'Levels', y = 'Probability', palette="mako")
plt.xlabel('Handicap Levels')
plt.ylabel('Probability')
plt.title('Posterior probability of Handicap given a no-show')
plt.show()
```

```python
sns.barplot(data = pd.DataFrame(posteriorNoShow('Gender')), 
            x = 'Levels', y = 'Probability', palette="mako")
plt.xlabel('Gender')
plt.ylabel('Probability')
plt.title('Posterior probability of Gender given a no-show')
plt.show()
```

```python
sns.barplot(data = pd.DataFrame(posteriorNoShow('DayofTheWeek')), 
            x = 'Levels', y = 'Probability', hue_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',
                       'Saturday', 'Sunday'], palette="mako")
plt.xlabel('DayofTheWeek')
plt.ylabel('Probability')
plt.title('Posterior probability of DayOfTheWeek given a no-show')
plt.show()
```

ECDF

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n
    return x, y

# Compute ECDFs
x_age, y_age = ecdf(df.Age)


# Plot all ECDFs on the same plot
plt.plot(x_age, y_age , marker='.', linestyle='none')

# Make nice margins
plt.margins(0.2)

# Annotate the plot
plt.legend(('setosa', 'versicolor', 'virginica'), loc='lower right')
plt.xlabel('petal length (cm)')
plt.ylabel('ECDF')
plt.xlim(-5,100)
# Display the plot
plt.show()
```

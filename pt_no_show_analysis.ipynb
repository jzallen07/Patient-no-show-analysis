In this project I will be taking a look into factors that migh be used to predict if a patient will show-up for their doctors appointment.

```python
#Import Packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab
import seaborn as sns
sns.set(style="whitegrid")
import datetime as dt
%matplotlib inline
```

```python
# Load the data

df = pd.read_csv('data/KaggleV2-May-2016.csv')
```

```python
# Examine the fields 

df.columns
```

From these headings some of them will need to be changed for spelling, clarity, and convenience. 

```python
# A few of the column names need to be corrected eithe due to spelling or convenience

df = df.rename(columns={'Hipertension':'Hypertension',
                        'No-show':'Noshow',
                        'Handcap':'Handicap'})
df.head()
```

```python
# The time series data need to be standardized for ease of use and we will create a day of the week column for later use

df['ScheduledDay'] = df['ScheduledDay'].apply(np.datetime64)
df['AppointmentDay'] = df['AppointmentDay'].apply(np.datetime64)

# Parse the day of the week from the appointment date

df['DayofTheWeek'] = df['AppointmentDay'].dt.weekday_name

print(df['ScheduledDay'].head())
print(df['AppointmentDay'].head())
print(df['DayofTheWeek'].head())
```

```python
# Exploring the oberview number of patients and appointments, as well as checking for missing values

print('number of patients:', '', df['PatientId'].nunique())
print('number of appointments:', '', df['AppointmentID'].nunique())
print('missing values?:', '\n', df.isnull().any())
```

Looks like we have a pretty complete set of data with a pretty good sample size.

```python
# Making the decision to exclude any age values that are either impossible of improbable

df = df[(df.Age >= 0) & (df.Age <= 100)]
```

I will be designating ages below zero as invalid data and those above 100 years of age (YOA) to be outliers. 

```python
df.describe()
```

## Data Wrangling

```python
# Taking a look at the age distribution

sns.distplot(df['Age'], bins=20)
plt.xlim(0,100)
```

A general trend that I might expect to see later once I investigate a little further is a lower number of men between the ages of 18 and around 60 showing up for their appointment or at the vary least fewer numbers of men even scheduling appointments as this is a general patient population trend from a public health level.

## Adding Variables

```python
# Waiting period calculation

df['WaitPeriod'] = (df['AppointmentDay'] - df['ScheduledDay']) / (np.timedelta64(1, 'D'))
```

It will be beneficial to examine how far ahead of the appointment date the appointment was scheduled. e.g. longer latency period might lead to higher no show rate. 

```python
# Taking a a different look at thing via gender

df_no = df[df.Noshow == 'No']
range_df = pd.DataFrame()
range_df['Age'] = range(100)

male = range_df.Age.apply(lambda x: len(df_no[(df_no.Age == x) & (df_no.Gender == 'M')]))
female = range_df.Age.apply(lambda x: len(df_no[(df_no.Age == x) & (df_no.Gender == 'F')]))

plt.plot(range(100),male, 'b')
plt.plot(range(100),female, color = 'g')
plt.legend(['M','F'])
plt.ylabel('Number of visits')
plt.xlabel('Age')
plt.title('Women visit the doctor more often')
plt.show()
```

```python
# Wait time men and women

plt.hold(True)
sns.kdeplot(df['WaitPeriod'][df.Gender == 'M'],marker = '*', color = 'b')
sns.kdeplot(df['WaitPeriod'][df.Gender == 'F'],marker ='o',color = 'g')
plt.xlim(-5, 50)
plt.title('Number of days spent waiting for an appoint')
plt.xlabel('Day before the appointment')
plt.legend(['Men','Women'])

plt.show()
```

From these plots we can see that there is a significant difference in the number appointments between men adn women. Further, there is NOT a significant difference between the amount of planning e.g. the number of days prior to attending the appointment that the appointment is scheudled between men and women.

# Associating factors


```python
def probbyCat(group_by):
    rows = []
    for item in group_by:
        for level in df[item].unique():
            row = {'Condition': item}
            total = len(df[df[item] == level])
            n = len(df[(df[item] == level) & (df['Noshow'] == 'No')])
            row.update({'Level': level, 'Probability': n / total})
            rows.append(row)
    return pd.DataFrame(rows)
```

```python
sns.barplot(data = probbyCat(['Diabetes', 'Alcoholism', 'Hypertension',
                                          'Scholarship']),
            x = 'Condition', y = 'Probability', hue = 'Level', palette="mako")
plt.title('Probability of showing up')

plt.show()
```

We can see some minor fluctuations in with medical conditions.

```python
sns.barplot(data = probbyCat(['Gender']),
            x = 'Condition', y = 'Probability', hue = 'Level', palette="mako")
plt.title('Probability of showing up')

plt.show()
```

Little or no direct fluctuation with gender.

```python
sns.barplot(data = probbyCat(['Handicap']),
            x = 'Condition', y = 'Probability', hue = 'Level', palette="mako")
plt.title('Probability of showing up')

plt.show()
```

Moderate and probably predictable changes in no show probability with increasing degree of handicap

```python
sns.barplot(data = probbyCat(['DayofTheWeek']),
            x = 'Condition', y = 'Probability', hue = 'Level', palette="mako",
           hue_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',
                       'Saturday', 'Sunday'])
plt.title('Probability of showing up')

plt.show()
```

Day of the week provides small amount of change as well.

```python
sns.barplot(data = probbyCat(['SMS_received']),
            x = 'Condition', y = 'Probability', hue = 'Level', palette="mako")
plt.title('Probability of showing up')

plt.show()
```
